```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Tsega\Scrapy\ico\ico\spiders\ico2.csv")
df
```

```python
# Define the mapping dictionaries for 'Category' and 'Platform'
category_mapping = {
    'Blockchain': 'Blockchain Service',
    'Artificial Intelligence': 'AI/VR',
    'DEX': 'DeFi',
    'Protocol': 'DeFi'
}

platform_mapping = {
    'ERC20': 'Ethereum',
    'BEP-20': 'Binance Chain',
    'Flow': 'Flow Network'
}

# Use the 'replace()' method to map the values in the 'Category' column using the 'category_mapping' dictionary
df['Category'] = df['Category'].replace(category_mapping)

# Use the 'replace()' method to map the values in the 'Platform' column using the 'platform_mapping' dictionary
df['Platform'] = df['Platform'].replace(platform_mapping)
```

```python
# Changing the ICO_period to the right format and renaming it to as the Start_date
def change_date(date):
    # Split the date string and take the first part before '–'
    new_date = date.split('–')[0].strip()
    # Concatenate '2023' to the new_date_str
    new_date += ' 2023'
    return pd.to_datetime(new_date, errors='coerce')


df['ICO_period'] = df['ICO_period'].apply(change_date)

df.rename(columns={'ICO_period' : 'Start_date'}, inplace=True)
```

```python
# Changing the Token_sale_end to the right format and renaming it to as the End_date
df['Token_sale_end'] = pd.to_datetime(df['Token_sale_end'])
df.rename(columns={'Token_sale_end' : 'End_date'}, inplace=True)
```

```python
# drop 'NOT SET'
df = df.drop(df[df['Raised_usd'] == 'NOT SET'].index)

# Remove the '$' and ',' and convert to an int
df['Raised_usd'] = df['Raised_usd'].str.replace('$', '').str.replace(',', '').astype(int)
```

```python
# Remove USD from the ICO_price and convert to a float
df['ICO_price'] = df['ICO_price'].str.replace(' USD', '').astype(float)
```

```python
# Remove rows where 'Fundraise_goal' contains 'Token'
df = df[~df['Fundraise_goal'].str.contains('Token')].reset_index(drop=True)

# Remove the '$' and ',' and convert to a float
df['Fundraise_goal'] = df['Fundraise_goal'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
# Renaming the Sold_%
df.rename(columns={'Sold_%' : 'Sold'}, inplace=True)

# Replace ',' with '.'  and '%'from the Sold column and convert to a float
df['Sold'] = df['Sold'].str.replace(',', '.').str.replace('%', '').astype(float)
```

```python
# Remove the ',' and convert to a float
df['Total_token'] = df['Total_token'].str.replace(',', '').astype(float)
```
