```python
# import pandas
import pandas as pd
```

```python
# load the scraped dataset
df = pd.read_csv(r"C:\Users\Tsega\Scrapy\ico\ico1.csv")
df
```

```python
# Convert start date to a date time
df['start_date'] = pd.to_datetime(df['start_date'])

# Convert end date to a date time
df['end_date'] = pd.to_datetime(df['end_date'])
```

```python
# Define the mapping dictionaries for 'Category', 'Token_Role', and 'Platform'
category_mapping = {
    'Blockchain Platform': 'Blockchain Service',
    'Blockchain': 'Blockchain Service',
    'Finance': 'DeFi',
    'Payments': 'DeFi',
    'Market':'Marketplace',
    'Marchket':'Marketplace',
    'Marketing': 'Marketing Service',
    'DEX': 'Exchange',
    'Verification': 'KYC/Security',
    'Advertising': 'Marketing Service',
    'Real Business': 'Business',
    'Social': 'Social Network',
    'Currency': 'DeFi',
    'Artificial Intelligence': 'AI/VR',
    'Predictions': 'Trading Platform',
    'Governance Token': 'DeFi',
    'Mobile': 'Business',
    'Media': 'Business',
    'Mining': 'Business',
    'Crowdfunding': 'Platform',
    'Insurance': 'DeFi',
    'Stablecoin': 'DeFi',
    'Smart Contract':'Platform',
    'SMarcht Contract':'Platform',
    'Wallet': 'DeFi',
    'Dapp': 'Platform',
    'Marchketplace': 'Marketplace',
    'Marchketing': 'Marketing Service',
    'VR': 'AI/VR',
    'KYC': 'KYC/Security',
    'Security': 'KYC/Security',
    'Cloud Storage': 'Cloud',
    'Cloud Computing': 'Cloud',
    'Lending': 'DeFi',
    'Investment': 'Platform',
    'Banking': 'DeFi',
    'Trading': 'Trading Platform',
    'Gambling': 'DeFi',
    'Marketing': 'Marketing Service',
    'Hybrid Intellingence':'AI/VR'
}

token_role_mapping = {
    'Utility Unsold Tokens': 'Utility',
    'Utility\nRegistered Company': 'Utility',
    'South Korea\nRole of Token': 'Utility',
    'Working MVP\nRole of Token': 'Utility',
    'Singapore\nRole of Token': 'Utility',
    '12\nTeam from': 'Utility',
    'USA\nRole of Token': 'Utility',
    '16\nTeam from': 'Utility',
    'Undefined': 'Utility'
}

platform_mapping = {
    'ERC20': 'Ethereum',
    'BEP-20': 'Binance Chain',
    'Flow': 'Flow Network',
    'BEP2': 'Binance Chain',
    'ERC721': 'Ethereum',
    'Matic Blockchain': 'Polygon',
    'ERC777': 'Ethereum',
    'ERC-1404': 'Ethereum',
    'ERC223': 'Ethereum',
    'TRC-10': 'TRON',
    'TRX-20': 'TRON',
    'WRC-20': 'WaykiChain',
    'WAN': 'Wanchain',
    'NEP-5': 'NEO Blockchain',
    'NRC20': 'Nebulas Chain',
    'RST-88': 'Realio Network',
    'OEP-4': 'Ontology Blockchain',
    'Ontology': 'Ontology Blockchain',
    'IRC-20': 'Iost Blockchain',
    'HECO': 'Heco Chain',
    'Own wallet': 'Own Blockchain',
    'Own blockchain': 'Own Blockchain'
}

# Use the 'replace()' method to map the values in the 'Category' column using the 'category_mapping' dictionary
df['category'] = df['category'].replace(category_mapping)

# Use the 'replace()' method to map the values in the 'Token_Role' column using the 'token_role_mapping' dictionary
df['token_role'] = df['token_role'].replace(token_role_mapping)

# Use the 'replace()' method to map the values in the 'Platform' column using the 'platform_mapping' dictionary
df['platform'] = df['platform'].replace(platform_mapping)
```

```python
# drop 'NOT SET'
df = df.drop(df[df['raised_usd'] == 'NOT SET'].index)

# Remove the '$' and ',' and convert to an float
df['raised_usd'] = df['raised_usd'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
# Remove USD, $ and the comma from the ICO_price and convert to a float
df['ico_price'] = df['ico_price'].str.replace(' USD', '').str.replace('$', '').str.replace(',', '').astype(float)
```

```python
# Remove the '$' and ',' and convert to a float
df['fundraise_goal'] = df['fundraise_goal'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
# Renaming the Sold_%
df.rename(columns={'sold_%' : 'sold'}, inplace=True)

df['sold'] = df['sold'].astype(float)
```

```python
# Remove the ',' and convert to a float
df['total_token'] = df['total_token'].str.replace(',', '').astype(float)

df
```

```python
df.info()
```

Transformation(Data Cleaning & Modelling)

```python
# Create a date dimension table
date = df[['start_date', 'end_date']].drop_duplicates().reset_index(drop=True)
date['start_day'] = date['start_date'].dt.day
date['start_month'] = date['start_date'].dt.month
date['start_year'] = date['start_date'].dt.year
date['start_weekday'] = date['start_date'].dt.weekday

date['end_day'] = date['end_date'].dt.day
date['end_month'] = date['end_date'].dt.month
date['end_year'] = date['end_date'].dt.year
date['end_weekday'] = date['end_date'].dt.weekday

# Assign the dateID with the index number and add "DA" prefix
date['date_id'] = 'DA' + date.index.astype(str)

# Rearrange the table
date = date[['date_id', 'start_date', 'start_year', 'start_month', 'start_weekday', 'start_day', 'end_date', 'end_year', 'end_month', 'end_weekday', 'end_day']]
date
```

```python
date.info()
```

```python
# Create a Project dimension table
project = df[['project', 'description', 'ticker']].drop_duplicates().reset_index(drop=True)

# Assign the PlatformID with the index number and add "PRO" prefix
project['project_id'] = 'PRO' + project.index.astype(str)

# Rearrange the table
project = project[['project_id', 'project', 'description', 'ticker']]
project
```

```python
project.info()
```

```python
# Create a Category dimension table
category = df[['category']].drop_duplicates().reset_index(drop=True)

# Assign the CategoryID with the index number and add "CAT" prefix
category['category_id'] = 'CAT' + category.index.astype(str)

# Rearrange the table
category = category[['category_id', 'category']]
category
```

```python
category.info()
```

```python
# Create a Platform dimension table
platform = df[['platform']].drop_duplicates().reset_index(drop=True)

# Assign the PlatformID with the index number and add "PLAT" prefix
platform['platform_id'] = 'PLAT' + platform.index.astype(str)

# Rearrange the table
platform = platform[['platform_id', 'platform']]
platform
```

```python
platform.info()
```

```python
# Create a token role dimension table
token_role = df[['token_role']].drop_duplicates().reset_index(drop=True)

# Assign the Token_roleID with the index number and add "TR" prefix
token_role['token_role_id'] = 'TR' + token_role.index.astype(str)

# Rearrange the table
token_role = token_role[['token_role_id', 'token_role']]
token_role
```

```python
token_role.info()
```

```python
# Create the fact tabele
icos = df.merge(project, on=['project', 'description', 'ticker']) \
        .merge(token_role, on='token_role') \
        .merge(platform, on='platform') \
        .merge(category, on='category') \
        .merge(date, on=['start_date', 'end_date'])

# Assign the PlatformID with the index number 
icos['ico_id'] = 'ICO' + icos.index.astype(str)

# Rearrange the fact table
icos = icos[['ico_id', 'date_id', 'project_id', 'platform_id', 'category_id', 'token_role_id', 'raised_usd', 'ico_price', 'fundraise_goal', 'total_token', 'sold']]
```

```python
# change raised usd to an int
icos['raised_usd'] = icos['raised_usd'].astype(float)

# change ico price to a float
icos['ico_price'] = icos['ico_price'].astype(float)

# change % sold to a float
icos['sold'] = icos['sold'].astype(float)

icos
```

```python
icos.info()
```

```python
# Convert all tables to a csv files
date.to_csv('date(DimT).csv', index = False)
project.to_csv('project(DimT).csv', index = False)
category.to_csv('category(DimT).csv', index = False)
platform.to_csv('platform(DimT).csv', index = False)
token_role.to_csv('token_role(DimT).csv', index = False)
icos.to_csv('icos(Fact Table).csv', index = False)
```
